```python
# for 변수 in range(횟수):
#   반복할 코드

for i in range(10):
    print("Hello World!", i)
```

```python
# 숫자를 감소시키기
for j in range(10, 0, -1):
    print("hello world", j)
```

```python
reversed(range(10))
```

```python
for b in reversed(range(10)):
    print("hello", b)
```

```python
# 입력한 횟수대로 반복하기

count = int(input('반복할 횟수를 입력하세요: '))

for i in range(count):
    print('Hello, World', i+1)
```

```python
# 시퀀스 객체로 반복하기
a = [10, 20, 30, 40, 50]
for i in a:
    print(i)
```

```python
fruits = ('apple', 'orange', 'grape')
for fruit in fruits :
    print(fruit)
```

```python
for letter in 'Python':
    print(letter, end = ' ')
```

```python
for letter in reversed("Python"):
    print(letter, end = '')
```

```python
# 16.5 연습문제: 리스트의 요소에 10을 곱해서 출력하기

x = [49, -17, 25, 102, 8, 62, 21]

for i in x:
    print(i * 10, end = ' ')
```

```python
# 16.6 심사문제: 구구단 출력하기

a = int(input())

for i in range(1,10):
    print(f"{a} * {i} =",a*i)
```
