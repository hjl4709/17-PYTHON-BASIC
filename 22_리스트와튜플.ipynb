```python
# while 문으로 요소 출력하기

a = [38, 21, 53, 62, 19]
i = 0

while i < len(a):
    print(a[i])
    i += 1
```

```python
# 리스트의 가장 작은 수, 가장 큰 수, 합계 구하기
# min(), max(), sum()

a = [38, 21, 53, 62, 19]
latgest = a[0]

for i in a:
    if i > latgest:
        latgest = i

print(a)
print(latgest)
print(min(a))
print(max(a))
```

```python
# 요소의 합계 구하기
a = [38, 21, 53, 62, 19]
x = 0

for i in a:
    x += i

print(x)
print(sum(a))
```

```python
# 리스트 표현식 사용하기
# [식 for 변수 in 리스트]
# list(식 for 변수 in 리스트)

a = [i for i in range(10)]

print(a)
```

```python
b = list(i for i in range(10))

print(b)
```

```python
c = [i + 5 for i in range(10)]

print(c)
```

```python
# 리스트 표현식에서 if 조건문 사용하기
# [식(변수) for 변수 in 리스트 if 조건식]
# list(식(변수) for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2 == 0] # 많이 사용!
print(a)

b = []

for i in range(10):
    if i % 2 == 0:
        b.append(i)
print(b)
```

```python
a = [i * j for j in range(2, 10) for i in range(1, 10)]
print(a)

b = []
for i in range(2, 10):
    for j in range(1, 10):
        b.append(i*j)
print(b)
```

```python
# 리스트에 map 사용하기 <<input 값이 있을 때는 항상 사용하기 때문에 매우 중요!>>

# list(map(함수, 리스트))
# tuple(map(함수, 튜플))

a = [1.2, 2.5, 3.7, 4.6]
for i in range(len(a)):
    a[i] = int(a[i])
print(a)

a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int, a))
print(a)
```
