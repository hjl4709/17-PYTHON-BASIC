```python
# 딕셔너리 조작하기
## setdefault, update

a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

a.setdefault("e") 
a.setdefault("f", 100) 
print(a)
```

```python
# 딕셔너리 키의 값 수정
a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

a.update(a=90)

print(a)
```

```python
a.update(a=900, f=60)

print(a)
```

```python
y = {
    1: "one",
    2: "two"
}

y.update({
    1: "ONE",
    3: "THREE"
})

print(y)
```

```python
y.update([
    [2, 'TWO'],
    [4, 'FOUR']
])

print(y)
```

```python
y.update(
    zip([1, 2], ['one', 'two'])
)

print(y)
```

```python
## 딕셔너리 표현식 사용하기

# {키: 값 for 키, 값 in 딕셔너리}
# dict({키: 값 for 키, 값 in 딕셔너리})

keys = ['a', 'b', 'c', 'd']
imsi = {
    'a': None,
    'b': None,
    'c': None,
    'd': None
}

# {for key, value in dict.fromkeys(keys).items()}
x = {a: b for a, b in imsi.items()}
print(x)
```

```python
print({key: 0 for key in dict.fromkeys(['a', 'b', 'c', 'd']).keys()})
print({value: 0 for value in {"a" : 10, "b" : 20, "c" : 30, "d" : 40}.values()})
print({value: key for key, value in {"a" : 10, "b" : 20, "c" : 30, "d" : 40}.items()})
```

```python
## 딕셔너리 표현식에서 if 조건문 사용하기
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

for key, value in x.items(): # -> RuntimeError: dictionary changed size during iteration 에러 발생 / 반복 중에 딕셔너리가 변경되었기 때문.
    if value == 20:
        del x[key]

print(x)        
```

```python
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

x = {key: value for key, value in x.items() if value != 20}
print(x)
```
