```python
## 재귀호출에 종료 조건 만들기

def hello(count):
    if count == 0:
        return
    print("Hello, world!", count)

    count -= 1
    hello(count)

hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
# 31.4 연습문제: 재귀호출로 회문 판별하기

def is_palindrome(word):
    if len(word) < 2:
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])

print(is_palindrome('hello'))
print(is_palindrome('level'))
```

```python
# 31.5 심사문제: 재귀호출로 피보나치 수 구하기

def fib(n):
    a = [0, 1, 2]
    for i in range(3, n):
        b = a[-1] + a[-2]
        a.append(b)
    return a[-1]


n = int(input())
print(fib(n))
```
