```python
# 나눗셈 후 소수점 이하를 버리는 // 연산자
print(3 // 2)
print(4 // 2.0)

# 나눗셈 후 나머지를 구하는 % 연산자
print(5 % 2)
print(5.0 % 2)
```

```python
# 거듭제곱을 구하는 ** 연산자
print(2 ** 10)
```
