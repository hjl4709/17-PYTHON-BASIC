```python
#계단식으로 별 출력하기 <<<시험문제 출제>>>
for i in range(5):
    for j in range(5):
        if j <= i:
            print("*", end='')
    print()
```

```python
# 대각선으로 별 출력하기

for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")            
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end="")
        else:
            print(" ", end='')
    print()
```

```python
# 19.5 연습문제: 역삼각형 모양으로 별 출력하기

for i in range(5):
    for j in range(5):
        if j < i:
            print(" ", end="")
        else:
            print("*", end="")
    print()
```

```python
# 19.6 연습문제: 산 모양으로 별 출력하기

a = int(input())

for i in range(a):
    for j in range(a*2 - 1):
        if a - 1 - i <= j <= a - 1 + i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
height = int(input())

for i in range(height):
    for j in range(height - (i+1)):
        print(" ", end="")
    for k in range(2 * i + 1):
        print("*", end="")
    print()
```

```python
height = int(input())

for i in range(1, height + 1):
    space = height - i
    star = 2 * i - 1
    print(" " * space + "*" * star + " " * space)
```
