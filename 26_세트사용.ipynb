```python
# 반복문으로 세트의 요소를 모두 출력하기

a = {1, 2, 3, 4}

for i in a:
    print(i)

for i in {1, 2, 3, 4}:
    print(i)
```

```python
# 세트의 표현식 사용하기

# {식 for 변수 in 반복가능한객체}
# set(식 for 변수 in 반복가능한객체)

a = {i for i in 'apple'}
print(a)
```

```python
## 세트 표현식에 if 조건문 사용하기

# {식 for 변수 in 세트 if 조건식}
# set(식 for 변수 in 세트 if 조건식)

a = {i for i in 'pineapple' if i not in 'apl'}

print(a)
```

```python
# 26.8 연습문제: 공배수 구하기

a = {i for i in range(1, 101) if i & 3 == 0}
b = {i for i in range(1, 101) if i & 5 == 0}

print(a & b)
```

```python
# 26.9 심사문제: 공약수 구하기

n, m = map(int, input().split())
a = {i for i in range(1, n+1) if n % i == 0}
b = {i for i in range(1, m+1) if n % i == 0}

divisor = a & b

result = 0
if type(divisor) == set:
    result = sum(divisor)

print(result)
```
