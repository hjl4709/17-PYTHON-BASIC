```python
# 변수의 사용범위 알아보기

x = 10

def foo():
    print(x)

foo()
print(x)
```

```python
def foo():
    a = 10 # 지역 변수
    print(a)

foo()
# print(a) # a는 함수 안에서 선언된 지역 변수 이므로 오류 발생
```

```python
# 함수 안에서 전역 변수 변경하기

b = 10
def foo():
    b = 20
    print(b)

foo()
print(b)
```

```python
c = 10
def foo():
    global c
    c = 20
    print(c)

foo()
print(c)
```

```python
def foo():
    global d
    d = 20
    print(d)

foo()
print(d)
```

```python
# 함수 안에서 함수 만들기

# def 함수이름():
#     코드
#     def 함수이름():
#         코드

def print_hello():
    hello = "Hello, world!"
    def print_message():
        print(hello)
    print_message()

print_hello()
```

```python
# 지역 변수 변경하기
def A():
    x = 10
    def B():
        x = 20

    B()
    print(x)

A()
```

```python
def A():
    x = 10
    def B():
        nonlocal x
        x = 20
    B()
    print(x)

A()
```

```python
# nonlocal이 변수를 찾는 순서

def A():
    x = 10
    y= 100

    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y
            x = x + 30
            y = y + 300
            print(x)
            print(y)
        C()
    B()
A()
```

```python
# global로 전역 변수 사용하기

x = 1
def A():
    def B():
        x = 20
        def C():
            global x
            x = x + 30
            print(x)
        C()
    B()
A()
```

```python
# 클로저 사용하기

def Calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b
    return mul_add

c = Calc() # 3x + 5 -> mul_add
print(c(1), c(2), c(3), c(4), c(5))
```

```python
# lambda로 클로저 만들기

def calc():
    a = 3
    b = 5
    return lambda x: a * x + b

c = calc()
print(c(1), c(2), c(3), c(4), c(5))
```

```python
# 33.5 연습문제: 호출 횟수를 세는 함수 만들기

def counter():
    i = 0
    def count():
        nonlocal i
        i += 1
        return i
    return count

c = counter()
for i in range(10):
    print(c(), end=' ')
```

```python
# 33.6 심사문제: 카운트다운 함수 만들기
def countdown(n):
    j = n + 1
    def count():
        nonlocal j
        j -= 1
        return j
    return count

n = int(input())

c = countdown(n)
for i in range(n):
    print(c(), end=' ')
```
