```python
## 서식 지정자로 소수점 표현하기
# "%f" % 숫자

"%f" % 2.3
```

```python
# "%.자릿수f" % 숫자
"%.1f" % 2.3
```

```python
"%10.3f" % 2.3
```

```python
## 서식지정자로 문자열 정렬하기
print("%10s" % 'python')
print("%-10s" % "python")
```

```python
## 서식지정자로 문자열 안에 값 여러개 넣기
"Today is %d %s" % (3, "April")
```

```python
## format 매서드 사용하기
"hello {0}".format('python') 
```

```python
"Hello {language} {version}".format(language="Python",version="3.14.6")
```

```python
## format 매서드로 문자열 정렬하기
print('{0:<10}'.format('phython'))
print('{0:>10}'.format('phython'))
print('{:>10}'.format('phython'))
```

```python
## 숫자 개수 맞추기
print('%03d' % 1)
print('{0:03d}'.format(35))
print('%08.2f' % 3.6)
print('{0:03.2f}'.format(3.6))
print('{0:08.2f}'.format(150.37))
```

```python
# 24. 6 심사문제: 높은 가격 순으로 출력하기
# 51900;83000;158000;367500;250000;59200;128500;1304000

price = list(map(int, input().split(";")))
price.sort(reverse=True)

for p in price:
    print('{:>9,}'.format(p))
```
