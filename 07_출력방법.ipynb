```python
# 값을 여러개 출력하기
print(1, 2, 3)

print("Hello", "Python")
```

```python
# sep로 값 사이에 문자 넣기

print(1, 2, 3, sep=", ")
print(4, 5, 6, sep="")
print(1920, 1080, sep="x")
```

```python
# 줄바꿈 활용하기
print(1, 2, 3)

print(1, 2, 3, sep="\n")

print(1, 2, 3,sep="\t")

print("1\n2\n3")
```

```python
# end 사용하기
print(1)
print(2)
print(3)

print(1, end=" ")
print(2, end=" ")
print(3, end=" ")
```

```python
# 7.3 퀴즈
# 1 : d
# 2 : c
# 3 : b, d, e

# 7.4 연습문제: 날짜와 시간 출력하기
year = 2000
month = 10
day = 27
hour = 11
minute = 43
second = 59

print(year, month, day, sep="/", end=" ")
print(hour, minute, second, sep=":")
```

```python
# 7.5 심사문제: 날짜와 시간 출력하기

year, month, day, hour, minute, second = input().split()
print(year, month, day, sep="-", end="T")
print(hour, minute, second, sep=":")
```
