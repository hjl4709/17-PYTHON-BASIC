```python
# else와 들여쓰기
x = 5
if x == 10:
    print("10입니다.")
else:
# print("x에 들어있는 숫자는") -> 들여쓰기가 잘못된 코드
    print("10이 아닙니다.")
```

```python
# if 조건문에 숫자 지정하기
## -> 숫자는 정수, 실수와 관계없이 0이면 거짓(False), 0이 아니면 참(True)이 나옴

if 0:
    print('참')
else:
    print('거짓')

if 1:
    print('참')
else:
    print('거짓')

if 0x1F:
    print('참')
else:
    print('거짓')

if 0b1000:
    print('참')
else:
    print('거짓')

if 13.5:
    print('참')
else:
    print('거짓')
```

```python
# if 조건문에 문자열 지정하기
## -> 문자열은 내용이 있으면 참, 빈 문자열은 거짓
if 'Hello':
    print('참')
else:
    print('거짓')

if '':
    print('참')
else:
    print('거짓')
```

```python
if not 0:
    print('참')

if not None:
    print('참')

if not '':
    print('참')
```

```python
# 조건식을 여러 개 지정하기

x = 10
y = 20

if (x == 10) and (y == 20):     # 언어에 따른 우선순위가 다르기 때문에 괄호로 조건을 묶어주는 습관이 좋음
    print('참')
else:
    print('거짓')
```

```python
# 14.7 심사문제: 합격 여부 판단하기

score = input().split()
everage = (int(score[0]) + int(score[1])+ int(score[2]) + int(score[3]))

if int(score[0]) > 100 or int(score[1]) > 100 or int(score[2]) > 100 or int(score[3]) > 100:
    print('잘못된 점수')
else:
    if everage >= 80:
        print("합격")
    else:
        print("불합격")
```

```python
score = list(map(int, input().split()))

t_score = 0

for s in score:
    if s < 0 or s > 100:
        print("잘못된 점수")
        break
    t_score += s
else:
    average = t_score / len(score)

    if average >= 80:
        print("합격")
    else:
        print("불합격")
```
