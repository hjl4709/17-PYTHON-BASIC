```python
# 부등호 사용하기
print(10 > 20)
print(10 < 20)
print(10 <= 10)
print(10 >= 10)
```

```python
# 객체가 같은지 다른지 비교하기

print(1 == 1.0)
print(1 is 1.0) # 사용하지 않는 방식, 다른 주소에 있음 -> False
print(1 is not 1.0)
```

```python
# 논리 연산자 사용하기

print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True or True)
print(True or False)
print(False or True)
print(False or False)

print(not True)
print(not False)
```

```python
print(not True and False or not False) # False and False => False or True => True
print(False and False or True)
```

```python
# 논리연산자와 비교연산자를 함께 사용하기
print(10 == 10 and 10 != 5)
print(10 > 5 or 10 < 3)
print(not 10 > 5)
print(not 1 is 1.0)
```
