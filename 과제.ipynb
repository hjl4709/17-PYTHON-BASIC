```python
# 문제 1

students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

print("====== 학생 성적 결과 =====")
t_score = 0
t_sub = 0

for name, scores in students.items():
    avg = sum(scores) / len(scores)
    t_score += sum(scores)
    t_sub += len(scores)

    print(f"이름: {name}")
    print(f"점수: {scores}")
    print(f"평균: {avg:.1f}\n")

t_avg = t_score / t_sub

print("====== 전체 결과 ======")
print(f"전체 학생 수: {len(students)}명")
print(f"전체 평균: {t_avg:.1f}")
```

```python
# 문제 10

col, row = map(int, input().split())
matrix = []

for i in range(row): 
    matrix.append(list(input()))

for r in range(row):
    for c in range(col):
        if matrix[r][c] == "*":
            continue

        count = 0
        for nr in range(r-1, r+2):
            for nc in range(c-1, c+2):
                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == "*":
                        count += 1
        matrix[r][c] = str(count)

for r in range(row):
    print("".join(matrix[r]))
```
