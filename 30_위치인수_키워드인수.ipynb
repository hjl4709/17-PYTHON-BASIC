```python
# 위치 인수와 리스트 언패킹
## 위치 인수를 사용하는 함수

def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10, 20, 30)
```

```python
## 언패킹 사용하기

x = [10, 20, 30]
print_numbers(*x)
```

```python
def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)

print_numbers(10, 20) # 오류발생 -> 매개변수보다 변수가 적기  때문
```

```python
## 가변 인수 함수 만들기

def print_numbers(*args):
    for arg in args:
        print(arg)
```

```python
print_numbers(10)
print_numbers(10, 20, 30, 40)
```

```python
x = [10]
y= [10, 20, 30, 40]

print_numbers(*x)
print_numbers(*y)
```

```python
# 키워드 인수와 딕셔너리 언패킹 사용하기
def personal_info(name, age, address):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)

x = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 용산구'
}

personal_info(**x)
# personal_info(name = '홍길동', age = '30', address = '서울시 용산구 이촌동') 둘이 같음
```

```python
# 키워드 인수를 사용하는 가변 인수 함수 만들기

def personal_info(**kwargs):
    # print(kwargs)
    # print(type(kwargs))
    for kw, arg in kwargs.items():
        print(kw, ': ', arg, sep='')
    print()

personal_info(name = '홍길동')
personal_info(name = '홍길동', age = '30', address = '서울시 용산구 이촌동')

x = {
    'name': '홍길동'
}

y = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 용산구'
}
personal_info(**x)
personal_info(**y)
```

```python
# 30.6 연습문제: 가장 높은 점수를 구하는 함수를 만들기

korean, english, mathematics, science = 100, 86, 81, 91

def get_max_score(*args):
    return max(args)

max_score = get_max_score(korean, english, mathematics, science)
print('높은 점수: ', max_score)

max_score = get_max_score(english, science)
print('높은 점수: ', max_score)
```

```python
# 30.7 심사문제: 가장 낮은 점수, 높은 점수와 평균 점수를 구하는 함수 만들기

korean, english, mathematics, science = map(int,input().split())

def get_min_man_score(*args):
    return min(args), max(args)

def get_average(korean = 0, english = 0, mathematics = 0, science = 0):
    return ((korean+english+mathematics+science)/ 4)

min_score, max_score = get_min_man_score(korean, english, mathematics, science)
average_score = get_average(korean=korean, english=english, mathematics=mathematics, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))

min_score, max_score = get_min_man_score(english,science)
average_score = get_average(english=english, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))
```

```python
korean, english, mathematics, science = map(int,input().split())

def get_min_man_score(*args):
    return min(args), max(args)

def get_average(korean=0, english=0,
                mathematics=0, science=0):
    a = [korean, english, mathematics, science]
    b = [i for i in a if i != 0]
    return sum(b) / len(b)

min_score, max_score = get_min_man_score(korean, english, mathematics, science)
average_score = get_average(korean=korean, english=english, mathematics=mathematics, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))

min_score, max_score = get_min_man_score(english,science)
average_score = get_average(english=english, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))
```
