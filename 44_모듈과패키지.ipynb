```python
# import로 모듈 가져오기

# import 모듈
# import 모듈1, 모듈2
# 모듈.변수
# 모듈.함수()
# 모듈.클래스()

import math

# math.pi
math.sqrt(4.0)
math.sqrt(2.0)
```

```python
## import as 로 모듈 이름 지정하기

import  math as m

print(m.sqrt(4.0))
print(m.pi)
```

```python
# from import 로 모듈의 일부만 가져오기

# import math

from math import pi, sqrt

print(pi)
sqrt(2.0)
```

```python
# from 모듈 import *

from math import *      # 부담이 되기 때문에 이렇게 사용은 잘 하지 않음

print(pi)
print(sqrt(2.0))
```

```python
# from import 로 모듈의 일부를 가져온 뒤 이름 지정하기

from math import sqrt as s, pi as p
print(s(2.0))
print(p)
```
