```python
# elif 사용하기

# if 조건식:
#     코드1
# elif 조건식:
#     코드2

x = 20

if x == 10:
    print("10입니다.")
elif x == 20:
    print("20입니다.")
```

```python
# if, elif, else 모두 사용

# if 조건식:
#     코드1
# elif 조건식:
#     코드2
# else:
#     코드3

x = 30
if x == 10:
    print("10입니다.")
elif x == 20:
    print("20입니다.")
else:
    print("10도 20도 아닙니다.")
```

```python
# 음료수 자판기 만들기

button = int(input())

if button == 1:
    print("콜라")
elif button == 2:
    print("사이다")
elif button == 3:
    print("환타")
else:
    print("제공하지 않는 메뉴")
```

```python
# 15.3 연습문제: if, elif, else 모두 사용하기

x = int(input())

if 11 <= x <= 20:
    print("11~20")
elif 21 <= x <= 30:
    print("21~30")
else:
    print("아무것도 해당하지 않음")
```

```python
# 15.4 심사문제: 교통카드 시스템 만들기

age = int(input())
balance = 9000

if 7 <= age <= 12:
    balance -= 650
elif 13 <= age <= 18:
    balance -= 1050
else:
    balance -= 1250

print(balance)
```
