```python
col, row = map(int, input().split())

# 1. 윗면 테두리 추가
matrix = [['.'] * (col + 2)]

# 2. 양옆 테두리를 추가하며 입력 받기
for _ in range(row):
    matrix.append(['.'] + list(input()) + ['.'])
    
# 3. 아랫면 테두리 추가
matrix.append(['.'] * (col + 2))

# 4. 이제 맵 밖으로 벗어날 일이 없으므로, 마음 편하게 8방향을 검사합니다.
# 인덱스가 1부터 시작함에 주의하세요.
for r in range(1, row + 1):
    for c in range(1, col + 1):
        if matrix[r][c] == '*':
            continue
            
        count = 0
        # 주변 8방향 탐색 (범위 체크 조건문이 완전히 사라짐!)
        for i in range(r - 1, r + 2):
            for j in range(c - 1, c + 2):
                if matrix[i][j] == '*':
                    count += 1
                    
        matrix[r][c] = str(count)

# 5. 출력할 때는 추가했던 테두리를 제외하고 출력합니다.
for r in range(1, row + 1):
    print("".join(matrix[r][1:col + 1]))
```
