```python
# while 반복문 사용하기

# 초기식
# while 조건식:
#     반복할 코드
#     변화식

i = 0
while i < 100:
    print("Hello, World!")
    i += 1
```

```python
# 초깃값을 1부터 시작하기
i = 1
while i <= 100:
    print("Hello, world!", i)
    i += 1
```

```python
i = 10
while i >0:
    print("Hello, world!", i)
    i -= 1
```

```python
# 입력한 횟수대로 반복하기

count = int(input('반복할 횟수를 입력하세요: '))

i = 0
while i < count:
    print('Hello, world!', i)
    i += 1
```

```python
# 17.5 연습문제: 변수 두 개를 다르게 반복하기
i = 2
j = 5

while i <= 32 or j >= 1:
    print(i, j)
    i *= 2
    j -= 1
```

```python
# 17.6 심사문제: 교통카드 잔액 출력하기
a = int(input())
while a >= 1350:
    print(a-1350)
    a -= 1350
```
